```python

import numpy as np
import matplotlib.pyplot as plt

fs = 100
maxt = 10
t = np.linspace(0, 10, 100)
amp = 5 # Amplitud
f = 50
x = amp * np.sin(2 * np.pi * f * t)

plt.plot(t, x, linewidth=3, label='Sine wave'); plt.grid()
plt.xlabel('tiempo [segundos]', fontsize=15)
plt.ylabel('Amplitud', fontsize=15)
plt.legend(fontsize=10, loc='upper right')
#plt.savefig('foo2.pdf')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
f = 60      # Frecuencia de la seal en Hz
fs = 100    # Frecuencia de muestreo en Hz
T = 0.4       # Duracin de la seal en segundos

# Vector de tiempo
t = np.arange(0, T, 1/fs)
print(t)

# Generar la seal senoidal
senoidal = np.sin(2 * np.pi * f * t)

# Graficar la seal
plt.plot(t, senoidal)
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.title("Seal Senoidal")
plt.grid()
plt.show()
```
