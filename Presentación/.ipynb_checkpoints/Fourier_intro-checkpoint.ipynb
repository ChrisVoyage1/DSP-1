# La Transformada Discreta de Fourier (DFT)

La transformada discreta de Fourier (DFT) convierte una secuencia finita de muestras equidistantes de una función en una secuencia de igual longitud de muestras equidistantes de la transformada de Fourier de tiempo discreto (DTFT), que es una función de frecuencia de valor complejo. El intervalo de muestreo de la DTFT es el inverso de la duración de la secuencia de entrada.

$$X[k] = \sum_{n=0}^{N-1} x[n] e^{-j\omega_kn}, \quad k=0,1,...,N-1$$
$$ \omega_k=\frac{2\pi}Nk, \quad f_k=\frac{f_s}Nk$$ 

con resolución $$\Delta=\frac{f_s}N$$

Vemos que X[n] no es sino el producto interno entre x[n] y s_k (factores de giro "twiddle factors")

$$ X[k] =\langle x,s_k \rangle = \sum_{n=0}^{N-1} x[n] s_k^*$$

$$ s_k[n] = e^{j\omega_kn}$$

a = 2 + j3

a* = 2 - j3



$$ \Omega = 2\pi f \quad \texttt{[rad/seg]}$$
$$ \omega = \Omega T = 2\pi f /f_s \quad \texttt{[rad/muestra]}$$
$$ F = \omega/\pi \quad \quad \texttt{[1/muestra] frecuencia normalizada}. $$

$$ X[k] =\langle x,s_k \ranglek = \sum_{n=0}^{N-1} x[n] s_k^*$$

```python
import numpy as np

def dft(x):
    N = len(x)
    X = np.zeros(N, dtype=complex)
    for k in range(N):  # Para cada frecuencia k
        for n in range(N):  # Para cada punto en la señal
            X[k] += x[n] * np.exp(-2j * np.pi * k * n / N)
    return X
    
señal = [1, 2, 3, 4]  # Señal de entrada
resultado = dft(señal)

print (np.round(resultado,decimals=1))
```

## DFT de cualquier senoidal compleja 

$$ x[n]=e^{j\omega_0n}, \quad \omega_0=\frac{2\pi}Nk_0 $$


$$X[k] = \sum_{n=0}^{N-1} e^{j\omega_0n} e^{-j\omega_kn}  = 
\sum_{n=0}^{N-1} e^{j\omega_0n} e^{-j\frac{2\pi}Nkn}, 
\quad k=0,1,...,N-1$$

$$X[k]= \frac{1-e^{-j2\pi(k-k_0)}}{1-e^{-j2\pi(k-k_0)/N}} $$
si k ≠ k_0, el denominador ≠ 0 y el numerador = 0

además X[k]=N para k=k_0 y X[k]=0 para k≠k_0

```python
plt.figure(1, figsize=(9.5, 7))

tol = 1e-5; N = 64; k0 = 7; 
X = np.array([]); x = np.exp(1j*2*np.pi*k0/N*np.arange(N))

plt.subplot(311)
plt.title('x; k_0 = 7, N = 64')
plt.plot(np.arange(N), np.real(x),'b', lw=1.5)
plt.plot(np.arange(N), np.imag(x),'g', lw=1.5)
plt.axis([0,N-1,-1,1])
for k in range(N):
	s = np.exp(1j*2*np.pi*k/N*np.arange(N))
	X = np.append(X, sum(x*np.conjugate(s)))

X.real[np.abs(X.real) < tol] = 0.0
X.imag[np.abs(X.imag) < tol] = 0.0

plt.subplot(312); plt.title('magnitude spectrum: abs(X)')
plt.plot(np.arange(N), abs(X), 'r', lw=1.5); plt.axis([0,N-1,0,N])

plt.subplot(313); plt.title('phase spectrum: angle(X)')
plt.plot(np.arange(N), np.angle(X),'c', lw=1.5); plt.axis([0,N-1,-np.pi,np.pi])

plt.tight_layout() #plt.savefig('dft-complex-sine-1.png') #plt.show()
```

## DFT de una senoidal real 

$$ x[n]=A_0 \cos{(\omega_0n)}= \frac{A_0}2 (e^{j\omega_0n}+e^{-j\omega_0n})
, \quad \omega_0=\frac{2\pi}Nk_0 $$


$$X[k] = \frac{A_0}2 \sum_{n=-N/2}^{N/2-1} (e^{j\omega_0n}+e^{-j\omega_0n}) e^{-j\omega_kn}$$

$$X[k]= \frac{A_0}2 \sum_{n=-N/2}^{N/2-1} e^{-j(\omega_0-\omega_k)n} +
\sum_{n=-N/2}^{N/2-1} e^{-j(\omega_0+\omega_k)n}$$

$$ X[k]=N \frac{A_0}2, k=k_0,-k_0 \quad X[k]=0, k \neq k_0,-k_0 $$

```python
plt.figure(1, figsize=(9.5, 7))
N = 64; k0 = 7
X = np.array([])
nv = np.arange(-N/2, N/2)
kv = np.arange(-N/2, N/2)
x = np.cos(2*np.pi*k0/N*nv)

plt.subplot(311)
plt.title('x; k_0 = 7.5, N = 64')
plt.plot(nv, x,'b', lw=1.5)
plt.axis([-N/2,N/2-1,-1,1])
for k in kv:
	s = np.exp(1j*2*np.pi*k/N*nv)
	X = np.append(X, sum(x*np.conjugate(s)))

X.real[np.abs(X.real) < tol] = 0.0; 
X.imag[np.abs(X.imag) < tol] = 0.0

plt.subplot(312);plt.title('magnitude spectrum: abs(X)')
plt.plot(kv, abs(X), 'r', lw=1.5); plt.axis([-N/2,N/2-1,0,N])

plt.subplot(313); plt.title('phase spectrum: angle(X)')
plt.plot(kv, np.angle(X),'c', lw=1.5); plt.axis([-N/2,N/2-1,-np.pi,np.pi])

plt.tight_layout()


```

# Propiedades de la DFT

## Linealidad
$$ a x_1[n] + b x_2[n] \xleftrightarrow{} a X_1[k] + b X_2[k]$$

$$ \begin{align*}
DFT\left(a x_1[n] + b x_2[n]\right)
&= \sum_{n=0}^{N-1} \left(a x_1[n] + b x_2[n]\right) e^{-j \frac{2\pi k n}{N}} \\
&= \sum_{n=0}^{N-1} a x_1[n] e^{-j \frac{2\pi k n}{N}} + \sum_{n=0}^{N-1} b x_2[n] e^{-j \frac{2\pi k n}{N}} \\
&= a \sum_{n=0}^{N-1} x_1[n] e^{-j \frac{2\pi k n}{N}} + b \sum_{n=0}^{N-1} x_2[n] e^{-j \frac{2\pi k n}{N}} \\
&= a X_1[k] + b X_2[k]
\end{align*} $$


```python
import numpy as np
import matplotlib.pyplot as plt

# Señales de ejemplo
N = 32  # Número de puntos
n = np.arange(N)
x1 = np.sin(2 * np.pi * 3 * n / N)  # Frecuencia 3
x2 = np.cos(2 * np.pi * 5 * n / N)  # Frecuencia 5

# Coeficientes escalares
a = 2; b = -1.5

# Combinación lineal en el dominio del tiempo
x = a * x1 + b * x2
# DFTs individuales
X1 = np.fft.fft(x1); X2 = np.fft.fft(x2)

# DFT combinada (usando linealidad)
X_lin = a * X1 + b * X2

# DFT real directamente de la señal combinada
X_real = np.fft.fft(x)

# Graficar
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.title("Señales originales en el dominio del tiempo")
plt.plot(n, x1, label="x1[n] = sin(2π·3·n/N)")
plt.plot(n, x2, label="x2[n] = cos(2π·5·n/N)")
plt.legend(); plt.grid()

plt.subplot(3, 1, 2)
plt.plot(n, x, color='purple', label="x[n] = a·x1[n] + b·x2[n]")
plt.title("Combinación lineal: x[n]"); plt.legend(); plt.grid()

# Graficar magnitudes de las DFTs
plt.subplot(3, 1, 3)
plt.title("Magnitud de la DFT - Comparación")
plt.plot(np.abs(X_real), label="DFT de x[n] (combinada)")
plt.plot(np.abs(X_lin), '--', label="a·DFT(x1[n]) + b·DFT(x2[n])")
plt.legend(); plt.grid()

plt.tight_layout(); plt.show()

# Graficar diferencia entre DFTs
plt.figure(figsize=(8, 3))
plt.title("Diferencia absoluta entre ambas DFTs")
plt.plot(np.abs(X_real - X_lin), color='red')
plt.grid(); plt.tight_layout(); plt.show()
```

## Traslación o corrimiento

$$ \begin{align*}
x[n-n_0] \iff e^{-j 2\pi k n_0 / N} X[k] \\
\text{DFT}(x[n-n_0])
&= \sum_{n=0}^{N-1-n_0} x[n-n_0] e^{-j 2\pi k n / N} \\
&= \sum_{m=-n_0}^{N-1-n_0} x[m] e^{-j 2\pi k (m+n_0) / N}, m=n-n_0 \\
&= \sum_{m=0}^{N-1} x[m] e^{-j 2\pi k m / N} e^{-j 2\pi k n_0 / N} \\
&= e^{-j 2\pi k n_0 / N} \sum_{m=0}^{N-1} x[m] e^{-j 2\pi k m / N} \\
&= e^{-j 2\pi k n_0 / N} X[k] \\
\end{align*} $$

Al desplazar la señal en el tiempo, su DFT sufre un cambio de fase lineal en frecuencia.

La magnitud de la DFT no cambia.

La fase sí cambia de manera predecible, y lo comprobamos comparando con la predicción teórica.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
N = 32
n = np.arange(N)
n0 = 5  # cantidad de muestras a desplazar

# Señal original
x = np.cos(2 * np.pi * 4 * n / N)  # frecuencia 4
x_shifted = np.roll(x, n0)         # desplazamiento circular a la derecha

# DFTs
X = np.fft.fft(x)
X_shifted = np.fft.fft(x_shifted)

# Fase teórica aplicada en frecuencia
k = np.arange(N)
phase_shift = np.exp(-1j * 2 * np.pi * k * n0 / N)
X_theoretical = X * phase_shift

# === GRAFICAMOS ===

plt.figure(figsize=(12, 8))

# Señales en el tiempo
plt.subplot(3, 1, 1)
plt.title("Señal original y desplazada (tiempo)")
plt.plot(n, x, label="x[n]")
plt.plot(n, x_shifted, '--', label=f"x[(n - {n0}) % N]")
plt.legend()
plt.grid()

# Magnitud de la DFT
plt.subplot(3, 1, 2)
plt.title("Magnitud de la DFT")
plt.plot(np.abs(X_shifted), label="DFT de señal desplazada")
plt.plot(np.abs(X_theoretical), '--', label="DFT teórica con fase")
plt.legend()
plt.grid()

# Fase de la DFT
plt.subplot(3, 1, 3)
plt.title("Fase de la DFT")
plt.plot(np.angle(X_shifted), label="Fase real")
plt.plot(np.angle(X_theoretical), '--', label="Fase teórica")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()

# Diferencia entre espectros
plt.figure(figsize=(8, 3))
plt.title("Diferencia absoluta entre DFT real y teórica")
plt.plot(np.abs(X_shifted - X_theoretical), color='red')
plt.grid()
plt.tight_layout()
plt.show()
```

## Simetría 
$$ \begin{align*}
x[n] \text{ real} \iff \Re \{ X[k] \} \text{ par e } \Im \{ X[k] \} \text{ impar}\\
\iff |X[k]| \text{ par e } \angle X[k] \text{ impar}\\
x[n] \text{ real y par} \iff \Re \{ X[k] \} \text{ par y } \Im \{ X[k] \} = 0\\
\iff |X[k]| \text{ par y } \angle X[k] = n\pi
\end{align*} $$

## Convolución

$$ \begin{align*}
\quad x_1[n] \ast x_2[n] \iff X_1[k] X_2[k] \\
\text{DFT}(x_1[n] \ast x_2[n])
&= \sum_{n=0}^{N-1} (x_1[n] \ast x_2[n]) e^{-j 2\pi k n / N} \\
&= \sum_{n=0}^{N-1} \sum_{m=0}^{N-1} x_1[m] x_2[n-m] e^{-j 2\pi k n / N} \\
&= \sum_{m=0}^{N-1} x_1[m] \sum_{n=0}^{N-1} x_2[n-m] e^{-j 2\pi k n / N} \\
&= \left( \sum_{m=0}^{N-1} x_1[m] e^{-j 2\pi k m / N} \right) X_2[k] \\
&= X_1[k] X_2[k] \\
\end{align*} $$

La convolución circular en el tiempo es equivalente al producto punto a punto en el dominio de la frecuencia.

Esta propiedad es la base para implementar convoluciones eficientes usando FFT en procesamiento de señales y deep learning.

La DFT no representa convolución lineal directamente, sino convolución circular. Si se desea una convolución lineal, se debe realizaar cero-padding antes de la DFT 

```python
import numpy as np
import matplotlib.pyplot as plt

# Longitud de la señal
N = 64
n = np.arange(N)

# Dos señales en el tiempo
x = np.sin(2 * np.pi * 3 * n / N)
h = np.cos(2 * np.pi * 5 * n / N)

# Convolución circular en el dominio del tiempo
x_circular_conv = np.fft.ifft(np.fft.fft(x) * np.fft.fft(h)).real

# Convolución circular usando np.roll y suma (por si quieres validar a mano)
# También puedes usar scipy.signal.fftconvolve con modo='same' si quieres más detalle

# DFTs individuales
X = np.fft.fft(x); H = np.fft.fft(h)

# Producto en frecuencia
Y = X * H

# Transformada inversa para obtener la convolución circular
y_from_freq = np.fft.ifft(Y).real

plt.figure(figsize=(12, 8))

# Señales originales
plt.subplot(3, 1, 1)
plt.title("Señales originales")
plt.plot(n, x, label="x[n] = sin(2π·3·n/N)")
plt.plot(n, h, label="h[n] = cos(2π·5·n/N)"); plt.legend(); plt.grid()

# Resultado de la convolución circular
plt.subplot(3, 1, 2)
plt.title("Convolución circular en el tiempo (DFT)")
plt.plot(n, x_circular_conv, label="x[n] ⊛ h[n]"); plt.grid(); plt.legend()

# Comparación entre tiempo y frecuencia
plt.subplot(3, 1, 3)
plt.title("Diferencia entre convolución desde tiempo vs frecuencia")
plt.plot(n, x_circular_conv - y_from_freq, label="Error")
plt.grid(); plt.legend(); plt.tight_layout()
plt.show()
```

## Conservación de la energía
$$ \sum_{n=-N/2}^{N/2-1} |x[n]|^2 = \frac{1}{N} \sum_{k=-N/2}^{N/2-1} |X[k]|^2 $$

```python
# import numpy as np
import matplotlib.pyplot as plt

# Señal de ejemplo: combinación de senos
N = 128*2  # Número de muestras
t = np.arange(N)
x = np.sin(2 * np.pi * 5 * t / N) + 0.5 * np.sin(2 * np.pi * 20 * t / N)
# Transformada Discreta de Fourier
X = np.fft.fft(x)
# Energía en el dominio del tiempo
energia_tiempo = np.sum(np.abs(x)**2)
# Energía en el dominio de la frecuencia (normalizada)
energia_frecuencia = (1 / N) * np.sum(np.abs(X)**2)

# Mostrar los resultados
print(f"Energía en el dominio del tiempo: {energia_tiempo:.6f}")
print(f"Energía en el dominio de la frecuencia: {energia_frecuencia:.6f}")

# Verificar que son aproximadamente iguales
if np.allclose(energia_tiempo, energia_frecuencia, atol=1e-10):
    print("✔ Se cumple la conservación de la energía (Parseval)")
else:
    print("✘ No se cumple la conservación de la energía")

# Espectro en dB (añadimos un valor pequeño para evitar log(0))
epsilon = 1e-12
X_dB = 20 * np.log10(np.abs(X) + epsilon)

# Graficar
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1); plt.plot(t, x)
plt.title('Señal en el dominio del tiempo')
plt.xlabel('n'); plt.ylabel('x[n]')

plt.subplot(1, 3, 2)
plt.plot(np.abs(X))
plt.title('Magnitud de la DFT')
plt.xlabel('k'); plt.ylabel('|X[k]|')

plt.subplot(1, 3, 3)
plt.plot(X_dB); #plt.xscale('log')
plt.title('Espectro en dB')
plt.xlabel('k'); plt.ylabel('Magnitud (dB)'); 

plt.tight_layout(); plt.show()
```

## Zero-Padding y convolución lineal

Zero padding significa agregar ceros al final (o a ambos lados) de una señal para aumentar su longitud.

Cuando haces convolución de dos señales de longitud L y M, el resultado debe tener longitud L + M - 1.

Pero la DFT es cíclica, así que si aplicamos DFTs de longitud menor que eso, se obtiene una convolución circular y el resultado será incorrecto (como si la señal "se doblara sobre sí misma"). La convolución lineal se puede hacer eficientemente con DFT si se hace zero-padding para evitar aliasing circular.

Sin padding:
La DFT multiplica y luego hace "wrap-around" del resultado (circular).

Con padding:
Si se hace padding para que ambas señales tengan longitud N ≥ L + M - 1, entonces la convolución circular coincide con la convolución lineal.

¿Qué es la convolución lineal?
Para dos señales x[n] de longitud L y h[n] de longitud M, su convolución lineal tiene longitud L + M - 1.

Pero si haces DFT de longitud N = L + M - 1 (o mayor), el resultado será correcto.


```python
import numpy as np
import matplotlib.pyplot as plt

# Señales más cortas para que se vea bien
x = np.array([1, 2, 3, 4])        # longitud L = 4
h = np.array([1, -1, 2])          # longitud M = 3
L = len(x); M = len(h)
N = L + M - 1                     # longitud necesaria para convolución lineal

# Convolución lineal directa (referencia)
linear_conv = np.convolve(x, h)

# Zero-padding hasta N
x_padded = np.pad(x, (0, N - L))
h_padded = np.pad(h, (0, N - M))

# DFTs
X = np.fft.fft(x_padded)
H = np.fft.fft(h_padded)

# Producto en frecuencia
Y = X * H

# Transformada inversa para obtener la convolución lineal
y_from_freq = np.fft.ifft(Y).real

# === GRAFICAMOS ===

plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.title("Convolución lineal: directa vs con DFT + padding")
plt.stem(linear_conv, linefmt='b-', markerfmt='bo', basefmt='k', label="np.convolve")
plt.stem(y_from_freq, linefmt='r--', markerfmt='ro', basefmt='k', label="DFT con padding")
plt.legend(); plt.grid()

plt.subplot(2, 1, 2)
plt.title("Diferencia entre ambas (error numérico)")
plt.stem(linear_conv - y_from_freq, linefmt='g-', markerfmt='go', basefmt='k'); plt.grid()

plt.tight_layout();plt.show()
```

## Zero-padding como interpolación en el dominio de Fourier

Objetivo:
Aumentar la resolución en frecuencia o interpolar la señal en el dominio del tiempo.

Y se realiza de forma muy simple:
Agregando ceros al final de la señal antes de hacer la FFT.

En el dominio de la frecuencia: ¡NO agrega información nueva! pero se interpola suavemente entre los puntos de frecuencia que ya existian.

Los puntos del espectro están más “juntos”, lo que da una visión más continua y detallada del espectro.

En el dominio del tiempo:
Si se hace la inversa (np.fft.ifft) de un espectro con ceros agregados en el medio, obtienes una versión interpolada de la señal original con más muestras.

Esto es como decir: “reconstruye esta señal como si tuviera una mayor tasa de muestreo”.

```python
import numpy as np
import matplotlib.pyplot as plt

# Señal original (una sinusoide corta)
N = 16
n = np.arange(N)
x = np.sin(2 * np.pi * 3 * n / N)

# Zero padding hasta N_pad (más resolución en frecuencia)
N_pad = 128
x_padded = np.pad(x, (0, N_pad - N))
X = np.fft.fft(x)
X_padded = np.fft.fft(x_padded)

# Frecuencias asociadas
f = np.fft.fftfreq(N, d=1)
f_padded = np.fft.fftfreq(N_pad, d=1)

# === Gráficas ===
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.title("Magnitud del espectro (sin padding)")
plt.plot(f, np.abs(X))
plt.grid()

plt.subplot(1, 2, 2)
plt.title("Magnitud del espectro (con zero padding)")
plt.plot(f_padded, np.abs(X_padded), 'r')
plt.grid()

plt.tight_layout()
plt.show()
```
